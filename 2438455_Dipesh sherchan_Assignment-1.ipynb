## Task- 3- Outlier Detection:

1. Identify outlier countries in South Asia based on their Score and GDP per Capita

```python
# Calculate Q1, Q3, and IQR for GDP per Capita and Score
Q1_gdp = southasia['Log GDP per capita'].quantile(0.25)
Q3_gdp = southasia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

Q1_score = southasia['score'].quantile(0.25)
Q3_score = southasia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Calculate outlier bounds
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

# Identify outliers
outliers_gdp = southasia[(southasia['Log GDP per capita'] < lower_bound_gdp) | (southasia['Log GDP per capita'] > upper_bound_gdp)]
outliers_score = southasia[(southasia['score'] < lower_bound_score) | (southasia['score'] > upper_bound_score)]

# Combine outliers from both columns
outliers = pd.concat([outliers_gdp, outliers_score]).drop_duplicates()

print("Outlier Countries:")
print(outliers[['Country name','score']])
```

 3. Metric Comparisons:

```python
# Compare metrics between South Asia and Middle East
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']

south_asia_means = southasia[metrics].mean()
middle_east_means = mideast[metrics].mean()

x = np.arange(len(metrics))  # Label locations
width = 0.35  # Width of bars

plt.figure(figsize=(10, 6))
plt.bar(x - width/2, south_asia_means, width, label='South Asia', color='purple')
plt.bar(x + width/2, middle_east_means, width, label='Middle East', color='cyan')

plt.xticks(x, metrics)
plt.ylabel('Mean Value')
plt.title('Comparison of Key Metrics between South Asia and Middle East')
plt.legend()
plt.show()

# Largest disparity
disparity = abs(south_asia_means - middle_east_means)
print("Metric showing the largest disparity:", disparity.idxmax())
```

4. Happiness Disparity:

```python
# Calculate range and coefficient of variation (CV) for each region
south_asia_range = southasia['score'].max() - southasia['score'].min()
middle_east_range = mideast['score'].max() - mideast['score'].min()

south_asia_cv = southasia['score'].std() / southasia['score'].mean()
middle_east_cv = mideast['score'].std() / mideast['score'].mean()

print(f"South Asia - Range: {south_asia_range:.2f}, CV: {south_asia_cv:.2f}")
print(f"Middle East - Range: {middle_east_range:.2f}, CV: {middle_east_cv:.2f}")

greater_variability = "South Asia" if south_asia_cv > middle_east_cv else "Middle East"
print(f"Region with greater variability in happiness: {greater_variability}")
```

Task 6: Outlier Detection

```python
# Detect outliers using the 1.5 * IQR rule
for region, data in [("South Asia", southasia), ("Middle East", mideast)]:
    Q1 = data['score'].quantile(0.25)
    Q3 = data['score'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data['score'] < lower_bound) | (data['score'] > upper_bound)]
    print(f"{region} Outliers:\n", outliers[['Country name', 'score']])
    print("\n")
```
