```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/drive/MyDrive/diabetes.csv')
```

```python
# Check for missing values
df.fillna(df.median(), inplace=True)

# Define features and target for regression
X_reg = df.drop(columns=["BloodPressure"])
y_reg = df["BloodPressure"]

# Split the data for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Linear Regression
reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)

# Predict and evaluate regression model
y_pred_reg = reg_model.predict(X_test_reg)
print("Regression Task:")
print(f"Mean Squared Error: {mean_squared_error(y_test_reg, y_pred_reg):.2f}")
print(f"R^2 Score: {r2_score(y_test_reg, y_pred_reg):.2f}\n")

# Define features and target for classification
X_clf = df.drop(columns=["Outcome"])
y_clf = df["Outcome"]

# Normalize features
scaler = StandardScaler()
X_clf = scaler.fit_transform(X_clf)

# Split the data for classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(X_clf, y_clf, test_size=0.2, random_state=42)

# Logistic Regression
log_reg = LogisticRegression()
log_reg.fit(X_train_clf, y_train_clf)
y_pred_clf_log = log_reg.predict(X_test_clf)

# Evaluate Logistic Regression
print("Classification Task - Logistic Regression:")
print(f"Accuracy: {accuracy_score(y_test_clf, y_pred_clf_log):.2f}")
print(f"Precision: {precision_score(y_test_clf, y_pred_clf_log):.2f}")
print(f"Recall: {recall_score(y_test_clf, y_pred_clf_log):.2f}")
print(f"F1 Score: {f1_score(y_test_clf, y_pred_clf_log):.2f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test_clf, y_pred_clf_log)}\n")

# K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_clf, y_train_clf)
y_pred_clf_knn = knn.predict(X_test_clf)

# Evaluate KNN
print("Classification Task - K-Nearest Neighbors:")
print(f"Accuracy: {accuracy_score(y_test_clf, y_pred_clf_knn):.2f}")
print(f"Precision: {precision_score(y_test_clf, y_pred_clf_knn):.2f}")
print(f"Recall: {recall_score(y_test_clf, y_pred_clf_knn):.2f}")
print(f"F1 Score: {f1_score(y_test_clf, y_pred_clf_knn):.2f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test_clf, y_pred_clf_knn)}")
```
