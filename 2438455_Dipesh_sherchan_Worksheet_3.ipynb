3.2 Exploratory Data Analysis Practice Exercise - 1.

Which passenger had the highest fare paid relative to their age?

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')
```

```python
df['fare_per_year']=df['Fare']/df['Age']
fare_per_year.head()
```

```python
high_fare_age = df[df['fare_per_year'] > 5]
high_fare_age.head()
```

```python
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
high_fare_age_srt.head()
```

```python
result = high_fare_age_srt[['Name', 'fare_per_year']]
result.head()
```

3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:

What percent of the total fare revenue came from each passenger class?

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')
```

```python
total_fare=df['Fare'].sum()
total_fare
```

```python
first_class_p=df[(df['Pclass'] == 1)]
total_fare_first_class_p=first_class_p['Fare'].sum()
total_fare_first_class_p
```

```python
second_class_p=df[(df['Pclass'] == 2)]
total_fare_second_class_p=second_class_p['Fare'].sum()
total_fare_second_class_p
```

```python
third_class_p=df[(df['Pclass'] == 3)]
total_fare_third_class_p=third_class_p['Fare'].sum()
total_fare_third_class_p
```

```python
combined_fare_list=[total_fare_first_class_p,total_fare_second_class_p,total_fare_third_class_p]
combined_fare_list
```

```python
percentage_first_class=(total_fare_first_class_p/total_fare)*100
percentage_second_class=(total_fare_second_class_p/total_fare)*100
percentage_third_class=(total_fare_third_class_p/total_fare)*100

print(f"Percentage of first class fare is {percentage_first_class} %")
print(f"Percentage of second class fare is {percentage_second_class} %")
print(f"Percentage of third class fare is {percentage_third_class} %")
```
